```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
from matplotlib.gridspec import GridSpec
import os

RESULTS = r"C:\Users\CyrilCorp\PyCharmMiscProject\ids-iot-combined\results"
plt.style.use('dark_background')

# ── CHART 1: Cross-Dataset F1 Grouped Bar Chart ───────────────────────────
models = ['Random\nForest', 'XGBoost', 'MLP', '1D-CNN', 'BiLSTM', 'Logistic\nRegression']
f1_iot  = [0.9872, 0.9805, 0.9786, 0.9780, 0.9779, 0.9744]
f1_iomt = [0.9995, 0.9991, 0.9979, 0.9974, 0.9978, 0.9942]

x = np.arange(len(models))
width = 0.35

fig, ax = plt.subplots(figsize=(14, 7))
fig.patch.set_facecolor('#1a0a2e')
ax.set_facecolor('#1a0a2e')

bars1 = ax.bar(x - width/2, f1_iot,  width, label='CIC-IoT-2023',  color='#7030A0', alpha=0.9, zorder=3)
bars2 = ax.bar(x + width/2, f1_iomt, width, label='CICIoMT-2024', color='#FFC446', alpha=0.9, zorder=3)

ax.set_ylim(0.96, 1.005)
ax.set_xticks(x)
ax.set_xticklabels(models, fontsize=13, color='white')
ax.set_ylabel('F1-Score', fontsize=14, color='white')
ax.set_title('F1-Score Comparison Across IoT Environments', fontsize=16, color='white', pad=15)
ax.tick_params(colors='white')
ax.spines['bottom'].set_color('#7030A0')
ax.spines['left'].set_color('#7030A0')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.yaxis.grid(True, color='#3a1a5a', zorder=0)
ax.set_axisbelow(True)

# Value labels on bars
for bar in bars1:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.0005,
            f'{bar.get_height():.4f}', ha='center', va='bottom', fontsize=9, color='#7030A0')
for bar in bars2:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.0005,
            f'{bar.get_height():.4f}', ha='center', va='bottom', fontsize=9, color='#FFC446')

# Gain annotations
for i, (a, b) in enumerate(zip(f1_iot, f1_iomt)):
    ax.annotate(f'+{b-a:.4f}', xy=(x[i]+width/2, b+0.002), ha='center',
                fontsize=9, color='#00cc66', fontweight='bold')

ax.legend(fontsize=13, facecolor='#1a0a2e', edgecolor='#7030A0', labelcolor='white')
plt.tight_layout()
plt.savefig(os.path.join(RESULTS, "chart_f1_grouped.png"), dpi=150, bbox_inches='tight', facecolor='#1a0a2e')
plt.show()
print("Saved: chart_f1_grouped.png")
```

```python
# ── CHART 3: SHAP Feature Importance Side by Side ─────────────────────────
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))
fig.patch.set_facecolor('#1a0a2e')

# CIC-IoT-2023 SHAP
features_iot = ['Number', 'ack_flag_number', 'HTTPS', 'IAT', 'Header_Length',
                'Tot size', 'Rate', 'ack_count', 'Tot sum', 'Max']
shap_iot = [0.1086, 0.0526, 0.0487, 0.0375, 0.0362, 0.0289, 0.0286, 0.0284, 0.0203, 0.0170]

ax1.set_facecolor('#1a0a2e')
bars = ax1.barh(features_iot[::-1], shap_iot[::-1], color='#7030A0', alpha=0.9)
ax1.set_xlabel('Mean |SHAP Value|', fontsize=12, color='white')
ax1.set_title('CIC-IoT-2023\nTop Features (Random Forest)', fontsize=13, color='white', pad=10)
ax1.tick_params(colors='white', labelsize=11)
ax1.spines['bottom'].set_color('#7030A0')
ax1.spines['left'].set_color('#7030A0')
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.xaxis.grid(True, color='#3a1a5a', zorder=0)
ax1.set_axisbelow(True)
for bar, val in zip(bars, shap_iot[::-1]):
    ax1.text(bar.get_width() + 0.001, bar.get_y() + bar.get_height()/2,
             f'{val:.4f}', va='center', fontsize=10, color='#FFC446')

# CICIoMT-2024 SHAP
features_iomt = ['rst_count', 'IAT', 'AVG', 'Tot size', 'ack_flag_number',
                 'Max', 'Header_Length', 'Number', 'Variance', 'Rate']
shap_iomt = [0.0568, 0.0522, 0.0522, 0.0452, 0.0380, 0.0329, 0.0315, 0.0304, 0.0300, 0.0253]

ax2.set_facecolor('#1a0a2e')
bars2 = ax2.barh(features_iomt[::-1], shap_iomt[::-1], color='#FFC446', alpha=0.9)
ax2.set_xlabel('Mean |SHAP Value|', fontsize=12, color='white')
ax2.set_title('CICIoMT-2024\nTop Features (Random Forest)', fontsize=13, color='white', pad=10)
ax2.tick_params(colors='white', labelsize=11)
ax2.spines['bottom'].set_color('#FFC446')
ax2.spines['left'].set_color('#FFC446')
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.xaxis.grid(True, color='#3a1a5a', zorder=0)
ax2.set_axisbelow(True)
for bar, val in zip(bars2, shap_iomt[::-1]):
    ax2.text(bar.get_width() + 0.001, bar.get_y() + bar.get_height()/2,
             f'{val:.4f}', va='center', fontsize=10, color='#7030A0')

plt.suptitle('SHAP Feature Importance Comparison', fontsize=16, color='white', y=1.01)
plt.tight_layout()
plt.savefig(os.path.join(RESULTS, "chart_shap_comparison.png"), dpi=150, bbox_inches='tight', facecolor='#1a0a2e')
plt.show()
print("Saved: chart_shap_comparison.png")
```

```python
# ── CHART 4: Model Performance Radar Chart (CIC-IoT-2023) ─────────────────
from matplotlib.patches import FancyArrowPatch
import matplotlib.patheffects as pe

categories = ['F1', 'AUC-ROC', 'Precision', 'Recall', 'Accuracy']
N = len(categories)
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

model_data = {
    'Random Forest':      [0.9872, 0.9952, 0.9989, 0.9758, 0.9760],
    'XGBoost':            [0.9805, 0.9958, 0.9622, 0.9994, 0.9622],
    'MLP':                [0.9786, 0.9929, 0.9999, 0.9582, 0.9601],
    'Logistic Regression':[0.9744, 0.9892, 1.0000, 0.9501, 0.9526],
}

colors = ['#FFC446', '#7030A0', '#00cc88', '#ff6b6b']

fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))
fig.patch.set_facecolor('#1a0a2e')
ax.set_facecolor('#1a0a2e')

for idx, (model, values) in enumerate(model_data.items()):
    vals = values + values[:1]
    ax.plot(angles, vals, 'o-', linewidth=2, color=colors[idx], label=model)
    ax.fill(angles, vals, alpha=0.1, color=colors[idx])

ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories, fontsize=13, color='white')
ax.set_ylim(0.93, 1.01)
ax.set_yticks([0.94, 0.96, 0.98, 1.00])
ax.set_yticklabels(['0.94', '0.96', '0.98', '1.00'], fontsize=9, color='#aaaaaa')
ax.grid(color='#3a1a5a', linewidth=0.8)
ax.spines['polar'].set_color('#7030A0')
ax.set_title('Model Performance Radar — CIC-IoT-2023', fontsize=15, color='white', pad=20)
ax.legend(loc='upper right', bbox_to_anchor=(1.35, 1.15), fontsize=11,
          facecolor='#1a0a2e', edgecolor='#7030A0', labelcolor='white')

plt.tight_layout()
plt.savefig(os.path.join(RESULTS, "chart_radar_ciciot.png"), dpi=150, bbox_inches='tight', facecolor='#1a0a2e')
plt.show()
print("Saved: chart_radar_ciciot.png")
```
